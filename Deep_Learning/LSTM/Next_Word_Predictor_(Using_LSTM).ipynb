First we will tokenize our data.

****What is Tokenization****
######A tokenizer is a tool in natural language processing that splits text into smaller units called tokens, such as words, subwords, or sentences, for easier analysis and processing.
Word tokenization specifically refers to dividing a text into individual words, typically by identifying spaces and punctuation marks as delimiters.

```python
import tensorflow as tf
from tensorflow.keras.preprocessing.text import Tokenizer
```

```python
tokenizer = Tokenizer()
```

The fit_on_texts() method takes a list of texts (strings) and updates the internal vocabulary based on the word frequency in these texts. It maps each unique word to a unique integer index, which can later be used to convert texts to sequences of these integer indices.

```python
tokenizer.fit_on_texts([RawData])
```

```python
#Lets check the total length of words in our data
len(tokenizer.word_index)
```

We process the raw text data by splitting it into sentences, tokenizing each sentence, and then generating all possible sub-sequences of tokens from each sentence. These sub-sequences can be used as training data for models that predict the next word in a sequence.

```python
input_sequences = []
for sentence in RawData.split('\n'):
  tokenized_sentence = tokenizer.texts_to_sequences([sentence])[0]

  for i in range(1,len(tokenized_sentence)):
    input_sequences.append(tokenized_sentence[:i+1])
```

```python
input_sequences
```

```python
max_len = max([len(x) for x in input_sequences])
```

**Apply padding On Our Tokenized Data**

The process of adjusting the lengths of sequences to a fixed size by adding a specific value (often zeros) to either the beginning or the end of each sequence, ensuring uniform input length for model training and evaluation.








```python
from tensorflow.keras.preprocessing.sequence import pad_sequences
padded_input_sequences = pad_sequences(input_sequences, maxlen = max_len, padding='pre')
```

```python
padded_input_sequences
```

```python
X = padded_input_sequences[:,:-1]
```

```python
y = padded_input_sequences[:,-1]
```

```python
X.shape
```

```python
y.shape
```

Each word is an output of the preceding word. So that's why we are converting them using one hot encoduing so our model can understand this data well for prediction. And our total length of words are 239 in our data.

```python
from tensorflow.keras.utils import to_categorical
y = to_categorical(y,num_classes=239)
```

```python
y.shape
```

##**LSTM MODEL**

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding, LSTM, Dense
```

**Embedded Layer**

Embebbed layer will create a dense vector of (100x100) for each tokenized word in our data. The parameter we are passing in our embedded layer are:

1.   Total no of words in our data which is 239
2.   The shape of embedded dense vectors are 100
3.   Our each row or sentence contains 29 words, So the input_length is 29.



**LSTM Layer**
###### We used 150 nodes of LSTM in our model. We tested that 150 nodes are working well for our model. We can modify it according to various problem

**Dense Layer**
###### We need to pass the exact number of words which we have in our data. We have 239 total words in our data. Because this layer will calculate the probability of eachw words and the tell us that word as output which will contains the highest probability.

```python
model = Sequential()
model.add(Embedding(239, 100, input_length=29))
model.add(LSTM(150))
model.add(Dense(239, activation='softmax'))
```

```python
model.compile(loss='categorical_crossentropy', optimizer='adam',metrics=['accuracy'])
```

```python
model.summary()
```

```python
model.fit(X,y,epochs=100)
```
